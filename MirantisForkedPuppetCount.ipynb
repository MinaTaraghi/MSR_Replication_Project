```python
import os
import pandas as pd

# this file computes puppet percentage for the forked Mirantis that are cloned on our local machine
# input: list of forked repos '/Users/baharannouri/Desktop/MSR_Project/forkedExel.xlsx'
# outpts: "forked_pp_percentage.csv" that contains the name and puppet oercentage of the forked mirantis repositories


df = pd.read_excel('/Users/baharannouri/Desktop/MSR_Project/forkedExel.xlsx') 
mylist = df['name'].tolist()

totalFiles = 0
totalDirectories = 0
totalppfile = 0

pp_percentage = []

for name in mylist:
    
    APP_FOLDER = '/Users/baharannouri/MSR_forked/{name}'.format(name=name)
    new_dict = {]}
    new_dict["name"] = name
    
    for base, dirs, files in os.walk(APP_FOLDER):
        
        for directories in dirs:
            totalDirectories += 1
        
        for Files in files:
            totalFiles += 1
            
            if Files.endswith(".pp"):
                totalppfile += 1
                #print(Files)
    

            
            
    print('FolderName:',name)
    print('FileNumber:',totalFiles)
    print('PPfiles:',totalppfile)
    
    print('ppPercentage:',(totalppfile/totalFiles)*100)
    new_dict["pp_percentage"] = totalppfile/totalFiles)*100
    pp_percentage.append(new_dict)

ppp = pd.DataFrame(pp_percentage)
ppp.to_csv("forked_pp_percentage.csv")


    
     
     
            

#print('Total number of files',totalFiles)
#print('Total Number of directories',totalDirectories)




```
